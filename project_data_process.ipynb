```python
import pandas as pd
```

```python
data = pd.read_csv("C:\sql_class\ElectronicsProductsPricingData.csv")
```

```python
data = data.drop_duplicates(subset=['id'])
data
```

```python
#drop the data with null value for manufacturer and shipping, drop the columns that are useless. 
df = data.dropna(subset = ['manufacturer'])[['id', 'prices.amountMax', 'prices.amountMin','prices.condition', 'prices.merchant','prices.shipping', 'brand', 'categories', 'dateAdded', 'dateUpdated', 'imageURLs', 'manufacturer', 'manufacturerNumber', 'name', 'sourceURLs']] 

df = df.dropna(subset = ['prices.shipping'])[['id', 'prices.amountMax', 'prices.amountMin','prices.condition', 'prices.merchant','prices.shipping', 'brand', 'categories', 'dateAdded', 'dateUpdated', 'imageURLs', 'manufacturer', 'manufacturerNumber', 'name', 'sourceURLs']] 
df
```

```python
df = df.rename(columns = {'prices.amountMax':'amountMax', 'prices.amountMin': 'amountMin', 'prices.merchant': 'merchant', 'prices.shipping': 'shipping'})
df.head()
```

```python

#get unique number of merchant, brand to generate random fake data.
merchant_count = df["prices.merchant"].unique().size
brand_count = df["brand"].unique().size
merchant_count
```

```python
brand_count
```

```python
#get the unique attributes, then combine with generated data.
uniqueBrand = pd.DataFrame(df.brand.unique(), columns = ['company'])
uniqueMerchant = pd.DataFrame(df.merchant.unique(), columns = ['merchant'])
```

```python
Brand_addtional = pd.read_csv("brand_additional.csv")
Merchant_addtional = pd.read_csv("merchant_additional.csv")
```

```python
brand = pd.concat([uniqueBrand, Brand_addtional], axis=1, join='inner')
Merchant = pd.concat([uniqueMerchant, Merchant_addtional], axis=1, join='inner')
```

```python
brand.head()
```

```python
#get manufacturer table
manufacturer = df[['manufacturer', 'manufacturerNumber']]
manufacturer.head()
```

```python
#split attributes that have multiple within one element
temp_image_df = df.imageURLs.str.split(',').apply(pd.Series, 1).stack()
```

```python
#name the new column
temp_image_df.name = 'image_url'
del df['imageURLs']
```

```python
# Drop index level to match main dataframe
temp_image_df.index = temp_image_df.index.droplevel(-1)
#join with main dataframe
df_image = df.join(temp_image_df)
df_image.head()
```

```python
#extract product_id and image_url, we get the image table
image = df_image[['id', 'image_url']]
image
```

```python
# do the same thing for category and scource url
temp_category_df = df.categories.str.split(',').apply(pd.Series, 1).stack()
temp_category_df.name = 'category'
del df['categories']
temp_category_df.index = temp_category_df.index.droplevel(-1)
#join with main dataframe
df_category = df.join(temp_category_df)
df_category.head()
```

```python
#extract product_id and category, we get the category table
category = df_category[['id', 'category']]
category
```

```python
temp_source_url_df = df.sourceURLs.str.split(',').apply(pd.Series, 1).stack()
temp_source_url_df.name = 'source_url'
del df['sourceURLs']
temp_source_url_df.index = temp_source_url_df.index.droplevel(-1)
#join with main dataframe
df_source_url = df.join(temp_source_url_df)
df_source_url.head()
```

```python
#extract product_id and category, we get the category table
#extract product_id and category, we get the category table
source_url = df_source_url[['id', 'source_url']]
source_url
```

```python
#get product table
product = df[['id', 'name', 'brand', 'merchant', 'shipping', 'amountMax', 'amountMin', 'dateAdded', 'dateUpdated', 'manufacturer']]
product.head()
```

```python
#rename attributes to match sql schema
product = product.rename(columns = {'id':'product_id', 'amountMax':'amountmax', 'amountMin' : 'amountmin', 'dateAdded' : 'date_added', 'dateUpdated' : 'date_updated'})
product.head()
category = category.rename(columns = {'id':'product_id'})
source_url = source_url.rename(columns = {'id':'product_id'})
image = image.rename(columns = {'id':'product_id'})
```

```python
#write dataframes to csv for future loading
product.to_csv('product.csv')
```

```python
image.to_csv('image.csv')
source_url.to_csv('source_url.csv')
category.to_csv('category.csv')
manufacturer.to_csv('manufacturer.csv')
brand.to_csv('brand.csv')
Merchant.to_csv('merchant.csv')
manufacturer.to_csv('manufacturer.csv')
```
